## Demos
The function ```show_in_notebook()``` should be called with ```cotab.table``` object, which holds the real data stored using astropy table class. When a method is applied to ```EdgeTable``` directly (i.e. ```cotab.show_in_notebook()```), it might not work since the way we implemented EdgeTable is different from the Astropy Table. However, in each EdgeTable object, it contains a variable called table, which holds the true data constructed using the Astropy table. That is, if any other functions that work fine for astropy table but do not work with the EdgeTable, then you could try applying the function to ```cotab.table```

````python
# this one will break
cotab.show_in_notebook()
````

````python
# this one should work fine
cotab.table.show_in_notebook()
````
